# Python 简介

Python是著名的“龟叔”Guido van Rossum在1989年圣诞节期间，为了打发无聊的圣诞节而编写的一个编程语言。

特点：面向对象，可升级，可靠扩展，可移植性，易学，易读，易维护。缺点：相对其他程序运行速度慢，可忽略；代码不能加密

## 文本编辑器

在Python的交互式命令行写程序，好处是一下就能得到结果，坏处是没法保存，下次还想运行的时候，还得再敲一遍。

所以，实际开发的时候，我们总是使用一个文本编辑器来写代码，写完了，保存为一个文件（.py），这样，程序就可以反复运行了。

Python文本编辑器：Sublime Text（免费使用，但是不付费会弹出提示框 http://www.sublimetext.com ）、 Notepad++（免费使用，有中文界面 https://notepad-plus-plus.org/ ）

使用Notepad++时注意：语言设置为Python，编程中选择“以UTF-8无BOM格式编码”
