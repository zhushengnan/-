循环 Loop

```python
line = '1 2 3 4 5'
fields = line.split()
fields
```

```python
total = 0
for field in fields:
    total += int(field)
total
```

```python
Python中有一种叫做列表推导式(List comprehension)的用法：
```

```python
numbers = [int(field) for field in fields]
numbers
```

```python
sum(numbers)
```

```python
写在一行：
```

```python
sum([int(field) for field in line.split()])
```
