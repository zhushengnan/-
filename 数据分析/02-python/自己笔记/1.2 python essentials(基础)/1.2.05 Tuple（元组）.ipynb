```python
不可变的tuple有什么意义？因为tuple不可变，所以代码更安全。如果可能，能用tuple代替list就尽量用tuple。
```

```python
t=()   #定义一个空的tuple
t
```

```python
t=(1)   #定义的不是tuple，是1这个数！这是因为括号()既可以表示tuple，又可以表示数学公式中的小括号，这就产生了歧义。
t       #因此，Python规定，这种情况下，按小括号进行计算，计算结果自然是1
```

```python
t=(1,)   #只有1个元素的tuple定义方式，必须加，
t
```

```python
“可变的”tuple
```

 
表面上看，tuple的元素确实变了，但其实变的不是tuple的元素，而是list的元素。

```python
l=['X','Y']
t=('a','b',l)
t
```

```python
l=['A','B']    #将l又赋值了，t元素不变，故t为原值
t
```

```python
l=['X','Y']     #将l元素进行了更改，t随之变化
t=('a','b',l)
l[0]='A'
l[1]='B'
t
```
