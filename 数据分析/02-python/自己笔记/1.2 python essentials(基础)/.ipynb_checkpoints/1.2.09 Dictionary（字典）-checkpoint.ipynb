```python
Python用{key:value}来生成Dictionary。
```

```python
d={'Michael':95,'Bob':75,'Tracy':85}
d['Michael']
```

```python
把数据放入dict的方法，除了初始化时指定外，还可以通过key放入
```

```python
d['Adam']=67   #一个key只能对应一个value，所以，多次对一个key放入value，后面的值会把前面的值冲掉
d['Adam']=88
d['Adam']
```

```python
'Thomas' in d   #key不存在，dict就会报错.可以通过in判断key是否存在
```

```python
d.get('Thomas')   #通过dict提供的get()方法，如果key不存在，可以返回None，返回None的时候Python的交互环境不显示结果
d.get('Thomas',-1)   #或者自己指定的value
```

```python
删除key
```

```python
d={'Michael':95,'Bob':75,'Tracy':85}   #pop(key)函数，同时对应的value值也一起删除
d.pop('Bob')
d
```
