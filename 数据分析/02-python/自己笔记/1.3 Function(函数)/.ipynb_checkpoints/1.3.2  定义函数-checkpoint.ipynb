## 定义函数

```python
def poly(x, a, b, c):                #用关键词def来定义函数。注意冒号：
    y = a * x ** 2 + b * x + c
    return y

x = 1
poly(x, 1, 2, 3)
```

```python
def my_abs(x):       #注意def+函数之后的冒号：
    if x>=0:         #注意if+条件后面的冒号：
        return x
    else:
        return -x
x=-1
my_abs(x)
```

```python
def nop():   #空函数，用pass语句完成.用来作为占位符
    pass     #比如现在还没想好怎么写函数的代码，就可以先放一个pass，让代码能运行起来。
```

```python
def my_abs(x): 
    if not isinstance(x,(int,float)):
        raise TypeError('bad operand type')#isinstance对参数类型做检查
    if x>=0:        
        return x
    else:
        return -x
my_abs('a')
```

```python
import math    #import math语句表示导入math包，并允许后续代码引用math包里的sin、cos等函数。
def move(x,y,step,angle=0):
    nx=x+step*math.cos(angle)
    ny=y-step*math.sin(angle)
    return nx,ny
x,y=move(100,100,60,math.pi/6)
print(x,y)   #在这里，Python函数返回的仍然是单一值
```

```python
r=move(100,100,60,math.pi/6)
print(r)    #返回值其实是一个tuple
```

```python
注意：定义函数时，需要确定函数名和参数个数；（def）

如果有必要，可以先对参数的数据类型做检查；（isinstance）

函数体内部可以用return随时返回函数结果；

函数执行完毕也没有return语句时，自动return None。

函数可以同时返回多个值，但其实就是一个tuple。
```
