```r
数值变量
```

```r
数值变量可进行加减乘除平均等运算，分为连续和离散变量。
```

```r
数据集中趋势的测量：均值（mean）、中位数(median)、众数（mode）
```

```r
数据分散趋势的测量：值域（range:max-min）、方差（variance）、标准差（standard variance）、四分位距（interquartile range）
```

```r
x<-c(1,9,2,8,3,9,4,5,7,6)
mean(x)
median(x)   #中位数
var(x)   #方差
sd(x)    #标准差
summary(x)  #最大、小值、四分位点
```

```r
一个变量的可视化：柱状图（histogram）、点图（dot plot）
```

```r
箱图（box plot）(中位数、分位点、极端值)
```

```r
两个变量的关系
```

```r
散点图（scatter plot）：方向、形状、强度、极端值
```
