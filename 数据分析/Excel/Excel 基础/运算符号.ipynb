一 、乘法运算。

1 简单的乘法运算。
2 复杂的数组运算。

例1：如下图所示，要求计算A产品的销售量。



公式:=SUMPRODUCT((B3:B9="A")\*C3:C9)

这里的*号其实还是乘的意思，只是一组数与另一组数相乘。B3:B9="A"的结果是一组逻辑值

{TRUE;FALSE;FALSE;FALSE;TRUE;TRUE;FALSE}

而C3:C9是一组数

{10;2;4;6;7;2;10}

它们两组数一对一相乘的结果是，数字\*TRUE=数字\*1，数字\*FALSE=数字\*0，它们相乘的目的也可以用一句话总结：
符合条件的保留，不符合条件的变成0，最后求和的结果就是符合条件的和。

二、通配符的应用

“\*”号作为通配符，表示任意多个字符（也包括0个）

2 在公式中的应用

=countif(a1:a10,"\*A\*") 统计包含A的单元格个数
=sumif(a1:a10,"\*A",b1:b10) 统计以A结尾的B列的和
=vlookup("\*a",b:d,3,0) 查找以a结束的...
=sum('\*'!a1) 可以跨表对所有表的A1单元格进行求和
