# 运算符

### AND & OR

AND 和 OR 可在 WHERE 子语句中把两个或多个条件结合起来。  
如果第一个条件和第二个条件都成立，则 AND 运算符显示一条记录。  
如果第一个条件和第二个条件中只要有一个成立，则 OR 运算符显示一条记录。  

```python
SELECT * FROM Persons WHERE FirstName='Thomas' AND LastName='Carter';                                #and且
SELECT * FROM Persons WHERE firstname='Thomas' OR lastname='Carter';                                 #or或
SELECT * FROM Persons WHERE (FirstName='Thomas' OR FirstName='William')AND LastName='Carter';        #and or结合使用
```

# 通配符

%

```python
SELECT * FROM Persons   #从上面的 "Persons" 表中选取居住在以 "N" 开始的城市里的人
WHERE City LIKE 'N%';
SELECT * FROM Persons      #包含 "lon"
WHERE City LIKE '%lon%;
```

_

```python
SELECT * FROM Persons               #选取名字的第一个字符之后是 "eorge" 的人
WHERE FirstName LIKE '_eorge';
SELECT * FROM Persons
WHERE LastName LIKE 'C_r_er';        #选取的这条记录的姓氏以 "C" 开头，然后是一个任意字符，然后是 "r"，然后是任意字符，然后是 "er"
```

[]

```python
SELECT * FROM Persons          #选取居住的城市以 "A" 或 "L" 或 "N" 开头的人
WHERE City LIKE '[ALN]%';
SELECT * FROM Persons
WHERE City LIKE '[!ALN]%';
```
